```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
data = pd.read_csv("PLTR.csv", index_col=0, parse_dates=True)
```

```python
data.head()
```

```python
exp1 = data["Close"].ewm(span=12, adjust=False).mean()
exp2= data["Close"].ewm(span=26, adjust=False).mean()
data["MACD"] = exp1 - exp2
data["Signal line"] = data["MACD"].ewm(span=9, adjust=False).mean()
```

```python
data.tail()
```

```python
fig, ax = plt.subplots()
data[['MACD', "Signal line"]].plot(ax=ax)
data["Close"].plot(ax=ax, alpha=0.25, secondary_y=True)
```

```python
high14 = data["High"].rolling(14).max()
low14 = data["Low"].rolling(14).min()
data["%K"] = (data["Close"] - low14)*100/(high14 - low14)
data['%D'] = data ['%K'].rolling(3).mean()
```

```python
data.tail()
```

```python

fig, ax = plt.subplots()

data[['%K','%D']].loc["2022-09-08":].plot(ax=ax)
ax.axhline(20, c="r", alpha=0.3)
ax.axhline(80, c="r", alpha=0.3)
data["Close"].loc["2022-09-08":].plot(ax=ax, alpha=0.3, secondary_y=True)
```
